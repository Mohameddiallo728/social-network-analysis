```python
# Importer les bibliothèques nécessaires
import networkx as nx
import matplotlib.pyplot as plt
import community
```

```python
# Créer un réseau social simple
G = nx.Graph()

# Ajouter des personnes (nœuds)
G = nx.Graph()
nodes = ["Alice", "Bob", "Charlie", "David", "Eve", "Frank", "Grace", "Hank", "Ivy", "Jack",
         "Kate", "Liam", "Mary", "Nancy", "Oliver", "Paul", "Quinn", "Rachel", "Steve", "Tina"]
G.add_nodes_from(nodes)

edges = [("Alice", "Bob"), ("Alice", "Charlie"), ("Bob", "David"), ("Bob", "Eve"),
         ("Charlie", "Frank"), ("Charlie", "Grace"), ("David", "Hank"), ("David", "Ivy"),
         ("Eve", "Jack"), ("Eve", "Kate"), ("Frank", "Liam"), ("Frank", "Mary"),
         ("Grace", "Nancy"), ("Grace", "Oliver"), ("Hank", "Paul"), ("Hank", "Quinn"),
         ("Ivy", "Rachel"), ("Ivy", "Steve"), ("Jack", "Tina")]
G.add_edges_from(edges)

# Visualiser le réseau social
pos = nx.spring_layout(G)  # positions pour tous les nœuds

# Dessiner les nœuds
nx.draw_networkx_nodes(G, pos, node_size=700)

# Dessiner les liens
nx.draw_networkx_edges(G, pos, width=2)

# Dessiner les étiquettes
nx.draw_networkx_labels(G, pos, font_size=10, font_family="sans-serif")

# Afficher le graphique
plt.title("Analyse de réseau social")
plt.axis("off")
plt.show()
```

Mesurons le nombre de connexions qu'une personne a

```python
# Centralité degré
centralite_degre = nx.degree_centrality(G)
centralite_degre
```

**Observation :** le noeud 1 (Alice) possède un  degré est 0.1053, ce qui signifie que Alice a environ 10,53% des connexions par rapport au nombre total de connexions possibles dans le réseau.
 Plus la centralité degré est élevée, plus le nœud est "central" dans le réseau, ce qui signifie qu'il a plus de connexions. Les nœuds 2 à 9 semblent avoir une centralité degré égale, indiquant qu'ils ont un nombre similaire de connexions.

 Identifions les personnes qui font le lien entre différentes parties du réseau.

```python
# Centralité d'intermédiarité
centralite_intermediarite = nx.betweenness_centrality(G)
centralite_intermediarite
```

**Analyse :**  prenons le nœud 2 (Bob) par exemple

La centralité d'intermédiarité est 0.6784, ce qui indique que Bob est souvent un point de passage clé entre d'autres nœuds(personne) dans le réseau. En d'autres termes, si tu veux atteindre rapidement un nœud(une personne) à partir d'un autre, le nœud 2(Bob) est souvent sur le chemin le plus court.

D'autre part, le nœud 11(Kate) a une centralité d'intermédiarité de 0, ce qui signifie qu'il n'est pas fréquemment utilisé comme un pont entre d'autres nœuds dans le réseau.

**Conclusion :** Les nœuds avec une centralité d'intermédiarité plus élevée sont souvent des "intermédiaires" importants dans le réseau, facilitant les connexions entre différentes parties.

Mesurons à quel point une personne est proche de toutes les autres.

```python
# Centralité de proximité
centralite_proximite = nx.closeness_centrality(G)
centralite_proximite
```

**Analyse :** prenons le nœud 2 par exemple

La centralité de proximité est 0.38, ce qui signifie que le nœud 2 est, en moyenne, à une distance de 0.38 du chemin le plus court vers tous les autres nœuds du réseau. En d'autres termes, le nœud 2 est assez proche des autres nœuds du réseau.

D'autre part, le nœud 20 a une centralité de proximité de 0.1979, ce qui indique qu'il est un peu plus éloigné des autres nœuds en moyenne par rapport au nœud 2.

**Conclusion :** Les nœuds avec une centralité de proximité plus élevée sont souvent plus "centraux" dans le réseau en termes de proximité avec d'autres nœuds.

```python
# Centralité degré
degree_centrality = nx.degree_centrality(G)

# Couleur des nœuds en fonction de la centralité degré
couleur_noeuds = list(degree_centrality.values())

# Visualiser le graphique avec des nœuds colorés
plt.figure(figsize=(10, 8))
nodes = nx.draw_networkx_nodes(G, pos, node_size=700, node_color=couleur_noeuds, cmap=plt.cm.Blues)
edges = nx.draw_networkx_edges(G, pos, width=2)
labels = nx.draw_networkx_labels(G, pos, font_size=10, font_family="sans-serif")

# Créer une barre de couleur (colorbar)
cbar = plt.colorbar(nodes, label="Centralité degré")
plt.title("Analyse de réseau social avec centralité degré")

# Afficher le graphique
plt.axis("off")
plt.show()
```

**Interpretation :**  Plus la couleur d'un nœud est foncée, plus sa centralité degré est élevée. Cela signifie que les nœuds plus foncés ont plus de connexions avec d'autres nœuds, et ils sont donc considérés comme plus "centraux" dans le réseau.

Les nœuds plus clairs, en revanche, ont moins de connexions et sont donc moins centraux dans le réseau.

Cette visualisation te donne une idée de la structure du réseau social en mettant en évidence les nœuds qui sont plus "importants" en termes de connexions. Cela peut être utile pour comprendre qui sont les acteurs clés dans le réseau en fonction de la centralité degré.

Detection si il ya une communauté qui se forme

```python
# Détecter les communautés à l'aide de la méthode de Louvain
partition = community.best_partition(G)

# Visualiser le graphe avec des nœuds colorés en fonction de la communauté
plt.figure(figsize=(10, 8))
node_colors = [partition[node] for node in G.nodes()]
cmap = plt.cm.get_cmap("viridis", max(node_colors) + 1)  # Choose a colormap
nx.draw_networkx_nodes(G, pos, node_size=700, node_color=node_colors, cmap=cmap)
nx.draw_networkx_edges(G, pos, width=2)
nx.draw_networkx_labels(G, pos, font_size=10, font_family="sans-serif")

# Afficher le graphique
plt.title("Analyse des réseaux sociaux avec détection des communautés")
plt.axis("off")
plt.show()
```
